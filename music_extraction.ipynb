```python
YOUTUBE_API_KEY = os.getenv('YOUTUBE_API_KEY')

# Create a YouTube API client
youtube = build('youtube', 'v3', developerKey=YOUTUBE_API_KEY)
```

```python

def get_youtube_video_url(search_query):
    # Search for the top 5 results for the song
    search_response = youtube.search().list(
        part="snippet",
        q=search_query,
        type="video",
        maxResults=5
    ).execute()

    video_ids = [item['id']['videoId'] for item in search_response['items']]
    video_titles = [item['snippet']['title'] for item in search_response['items']]

    # Get content details (duration) for each video
    details_response = youtube.videos().list(
        part="contentDetails,snippet",
        id=','.join(video_ids)
    ).execute()

    video_data = []

    for item in details_response['items']:
        video_id = item['id']
        title = item['snippet']['title']
        duration = item['contentDetails']['duration']  # ISO 8601 duration
        video_url = f"https://www.youtube.com/watch?v={video_id}"
        video_data.append({
            'title': title,
            'duration': duration,
            'url': video_url
        })

    return video_data

# Function to convert ISO 8601 duration to seconds
def duration_to_seconds(duration):
    import isodate
    parsed_duration = isodate.parse_duration(duration)
    return int(parsed_duration.total_seconds())

def find_best_video(video_data):
    # Convert all durations to seconds
    durations = [duration_to_seconds(video['duration']) for video in video_data]

    # Find the most common duration
    common_duration = Counter(durations).most_common(1)[0][0]

    # First try to find a video with 'audio' or 'lyrics' in the title, within the common duration
    for video in video_data:
        if ('audio' in video['title'].lower() or 'lyrics' in video['title'].lower()) and duration_to_seconds(video['duration']) == common_duration:
            return video

    # If no suitable title, return a video that matches the common duration
    for video in video_data:
        if duration_to_seconds(video['duration']) == common_duration:
            return video

    # If no matches at all, return the first result (as a fallback)
    return video_data[0]
```

```python
# Testing to see if functions work
# Example: Search for "Die With A Smile by Lady Gaga Official Audio"
search_query = "Die With A Smile by Lady Gaga Official Audio"
video_data = get_youtube_video_url(search_query)

# Find the best video based on time and title criteria
best_video = find_best_video(video_data)

print(f"Best Video Title: {best_video['title']}")
print(f"Best Video URL: {best_video['url']}")
```

```python
for genre, song_list in genre_songs_dict.items():

    # Store tuples of song name and URL
    url_list = []

    for song in song_list:
        search_query = (f"{song} Official Audio")

        try:
            video_data = get_youtube_video_url(search_query)
            best_video = find_best_video(video_data)
            if best_video:
                url_list.append((song, best_video['url']))
                print(f"Best Video Title: {best_video['title']}")
                print(f"Best Video URL: {best_video['url']}\n")
            else:
                print(f"No suitable video found for {song}")
        except Exception as e:
            print(f"Error processing {song}: {str(e)}")


    youtube_song_urls[genre] = url_list
```
